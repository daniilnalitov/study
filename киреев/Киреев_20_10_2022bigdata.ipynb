1. Создать (используя модуль numpy и pandas) три файла *.csv, содержащих 5 полей:
дата, id, пол, профессия, доход. В каждом из файлов используя 1000 уникальных id,
4 уникальных профессии, 2 пола, 3 года с гранулярностью день, получить 1 млн. строк.

```python
import numpy as np
import pandas as pd
```

```python
df0=pd.DataFrame({'sex':np.random.choice(['муж','жен'],p=(0.65,0.35),size=1000),
'id':np.array(range(1000))})
df0[:5]
```

```python
df0['profession']=[x for x in np.random.choice(['специалист','старший специалист','руководитель отдела','директор'],size=(1000,3000))]
df0[:5]
```

```python
df1=df0.explode('profession')
df1['DATE']=list(pd.date_range('2010-01-01',periods=3000,freq='D'))*1000
df1['income']=list(np.random.normal(100000,20000,size=3000))*1000
df1
```

```python
del df0
```

```python
df2=df1.sample(1000000)
df2.to_csv('/content/df1.csv',sep=';',index=False,mode='w')
```

```python
df1.index=range(3000000)
df1
```

```python
df3=df1[~df1.index.isin(df2.index)].sample(1000000)
df3.to_csv('/content/df2.csv',sep=';',index=False,mode='w')
```

```python
df4=df1[~((df1.index.isin(df2.index))|(df1.index.isin(df3.index)))]
df4.to_csv('/content/df3.csv',sep=';',index=False,mode='w')
```

```python
dct=dict()
for path in pathl:
  with open(path) as fl:
    for line in fl.readlines()[1:]:
       #print(line)
       key=line.split(';')[0]+''.join(line.split(';')[3].split('-')[:2])
       value=float(line.split(';')[-1])
       try:
         dct[key]=[{'sum':dct[key][0]['sum']+value},{'cnt':dct[key][1]['cnt']+1}]
       except:
         dct.update({key:[{'sum':value},{'cnt':1}]})
dct
```

```python
dct['жен201001']
```

```python
df=pd.DataFrame()
for path in pathl:
  for chunk in pd.read_csv(path,sep=';',chunksize=100000):
    tmp=chunk.groupby([chunk.sex+chunk.DATE.apply(lambda x: x.split('-')[0]+x.split('-')[1])])['income'].agg({'sum','count'})
    df=df.append(tmp.reset_index())
df.groupby(['index']).sum()
```

```python
df.groupby([df.sex+df.DATE.apply(lambda x: x.split('-')[0]+x.split('-')[1])])['income'].agg({'sum','count'}).reset_index()
```

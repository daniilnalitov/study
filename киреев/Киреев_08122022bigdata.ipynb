```python
import time
```

```python
links=[]
for i in range(10):
  url='https://rosatom.ru//journalist/arkhiv-novostey/?from=01.01.2022+00%3A00%3A00&to=31.12.2022+23%3A59%3A59&AJAX=Y&PAGEN_1={}&AJAX=Y'.format(str(i))
  wd = webdriver.Chrome('chromedriver',options=chrome_options)
  wd.get(url)
  time.sleep(10)
  wd.refresh()
  pg=wd.page_source
  links+=BS(pg).find_all('a')
```
